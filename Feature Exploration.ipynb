```python
import pandas as pd
from glob import glob
from PIL import Image
from bokeh.charts import Bar, show, output_notebook

%matplotlib inline
names = {0:'airplane', 1:'automobile', 2:'bird', 3:'cat', 4:'deer', 5:'dog', 6:'frog', 7:'horse', 8:'ship', 9:'truck'}
```

```python
labels = pd.read_csv('./Data/train/trainLabels.csv', header=None)
labels.columns = ['File','Class']
labels = labels.groupby(['Class']).size().reset_index()
labels.columns = ['Class','Images in Class']
labels['Class'] = labels['Class'].apply(lambda x: names[x])
b = Bar(labels, label='Class', values='Images in Class', legend='bottom_left', title='Class Distribution of Training Images')
output_notebook()
show(b)
```
